```python
import pandas as pd
import itertools
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score
```

```python
loss = ["linear", "square", "exponential"]
n_estimators = [50,100,150,200]
```

```python
results = []
```

```python
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset = dataset.astype(int)
dataset 
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
independent
```

```python
dependent = dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent,test_size = 0.30, random_state = 0)
x_train
```

```python
for loss, n_estimators in itertools.product(loss, n_estimators):
    try:
        # Create and train the model
        model = AdaBoostRegressor(
            loss=loss,
            n_estimators=n_estimators,
            random_state=0
        )
        model.fit(x_train, y_train)

        # Predict and calculate R² score
        y_pred = model.predict(x_test)
        score = r2_score(y_test, y_pred)

        # Store result
        results.append({
           "loss" : loss,
            "n_estimators" : n_estimators,
            "r2_score": score
        })

    except Exception as e:
        # Log errors gracefully (e.g., for unsupported combinations)
        results.append({
            "loss": loss,
            "n_estimators" : n_estimators,
            "r2_score": None,
            "error": str(e)
        })
```

```python
df_results = pd.DataFrame(results)

# Save to Excel
df_results.to_excel("AdaBoostResult.xlsx", index=False)

print("✅ Results saved to 'AdaBoostResult.xlsx'")
```
